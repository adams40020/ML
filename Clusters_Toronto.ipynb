```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_toronto = pd.read_csv('Toronto_neighbourhood.csv')
```

```python
df_toronto.head()
```

```python
df_with_toronto = df_toronto[df_toronto['Borough'].str.contains('Toronto')]
df_with_toronto.reset_index()
df_with_toronto.drop(columns = ['Unnamed: 0','Postcode'],inplace=True)
```

```python
df_with_toronto
```

```python
from sklearn.cluster import KMeans
import folium
```

```python
map = folium.Map(location=[43, -80], zoom_start=9)
```

```python
for i in range(0,len(df_with_toronto)):
    folium.Circle([df_with_toronto.iloc[i]['Latitude'], df_with_toronto.iloc[i]['Longitude']],popup=df_with_toronto.iloc[i]['Borough'],
                        color='blue',fill=True,fill_color='blue',fill_opacity=0.7,radius=0).add_to(map)
map
```

```python
# set number of clusters
df_clusters = df_with_toronto[['Latitude','Longitude']]
kclusters = 5

# run k-means clustering
kmeans = KMeans(n_clusters=3, random_state=0).fit(df_clusters)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
df_clusters.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df_clusters
```

```python
#Clusters are now differented by size, cluster 0 has lowest radies, 1 is greater than 0, 2 has highest size
map2 = folium.Map(location=[43, -80], zoom_start=9)
for i in range(0,len(df_clusters)):
   folium.Circle(
      location=[df_clusters.iloc[i]['Latitude'], df_clusters.iloc[i]['Longitude']],
      radius=df_clusters.iloc[i]['Cluster Labels']*200,
      color='crimson',
      fill=True,
      fill_color='crimson'
   ).add_to(map2)
map2
```
